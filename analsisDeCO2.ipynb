```python
plt.scatter(df['ENGINESIZE'], df['FUELCONSUMPTION_COMB_MPG'], color='blue')
plt.title('Fuel consumption')
plt.xlabel('Enginesize')
plt.ylabel('Fuel consumption')
plt.show()
```

```python
plt.scatter(df['CYLINDERS'], df['FUELCONSUMPTION_COMB_MPG'], color='blue')
plt.title('Fuel consumption')
plt.xlabel('Cylinders')
plt.ylabel('Fuel consumption')
plt.show()
```

```python
plt.scatter(df['CYLINDERS'], df['CO2EMISSIONS'], color='blue')
plt.title('Emissions vs cylinders')
plt.xlabel('Cylinders')
plt.ylabel('CO2 emissions')
plt.show()
```

```python
plt.scatter(df['ENGINESIZE'], df['CO2EMISSIONS'], color='blue')
plt.title('Emissions vs engine size')
plt.xlabel('Engine size')
plt.ylabel('CO2 emissions')
plt.show()
```
